```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Load the UC San Diego GPA data
UCSD_GPA = pd.read_csv('sdGPA.csv')  # Replace 'sdGPA.csv' with the correct path if needed

# Check the column names to ensure 'Fall Term' exists
print(UCSD_GPA.columns)

# Filter data for 2018 and later (make sure 'Fall Term' column exists)
UCSD_GPA = UCSD_GPA[UCSD_GPA['Fall Term'] >= 2018]

# Sort data by 'Fall Term'
UCSD_GPA = UCSD_GPA.sort_values('Fall Term')

# Display the filtered and sorted data
print(UCSD_GPA)
```

```python
import pandas as pd

# Data for UCSD Freshman Enrollee GPA Count
data = {
    'Fall Term': [2023, 2022, 2021, 2020, 2019, 2018],
    '4.0 and above': [5778, 5456, 5962, 4753, 4147, 4700],
    '3.7 - 3.9': [882, 760, 1101, 1153, 1213, 1388],
    '3.30 - 3.69': [98, 95, 163, 290, 337, 337],
    '3.00 - 3.29': [0, 0, 10, 0, 16, 0],
    '0.00 - 2.99': [0, 0, 7, 0, 5, 0],
    'Unknown - Missing': [247, 229, 300, 240, 305, 262]
}

# Create and display the DataFrame
UCSD_GPA = pd.DataFrame(data)
UCSD_GPA.fillna(0, inplace=True)

UCSD_GPA
```

```python
# Create the line plots for each GPA range
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='4.0 and above', label='4.0 and above')
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='3.7 - 3.9', label='3.7 - 3.9')
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='3.30 - 3.69', label='3.30 - 3.69')
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='3.00 - 3.29', label='3.00 - 3.29')
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='0.00 - 2.99', label='0.00 - 2.99')
sns.lineplot(data=UCSD_GPA, x='Fall Term', y='Unknown - Missing', label='Unknown - Missing')

# Relabel the graph
plt.xlabel('Fall Term')
plt.ylabel('# of students')
plt.title("UCSD Freshman Enrollees' GPA by Count")

# Remove y-axis tick labels
plt.yticks([])

# Add a legend and move it outside of the graph for visibility
plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Display the plot
plt.show()
```
